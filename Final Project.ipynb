## **______________________________________ SALES IN RETAIL STORE ______________________________________**

######

 The dataset used in this project, **“Retail Store Sales: Dirty for Data Cleaning,”** contains retail transaction details with common real-world data
 problems such as missing values, incorrect entries, and duplicates. Each row represents a single transaction, and the dataset includes information
 such as:

>Transaction ID (object):- A unique identifier for each transaction.

>Customer ID (object):- Identifies which customer made the purchase.

>Category (object):- The product category, such as electronics, clothing, etc.

>Item (object):- The specific product purchased.

>Price Per Unit (float64):- The cost of a single unit of the item.

>Quantity (float64):- The number of units bought in the transaction.

>Total Spent (float64):- The total amount spent (Price × Quantity).

>Payment Method (object):- Mode of payment used, e.g., Cash, Credit Card, Digital Wallet.

>Location (object):- Whether the purchase was made in-store or online.

>Transaction Date (object):- The date on which the transaction occurred.

>Discount Applied (object):- Indicates if a discount was applied to the purchase.

Since the dataset is intentionally “dirty,” it contains missing values, duplicates, and inconsistent data, making it suitable for practicing data cleaning and analysis. This dataset helps in understanding sales patterns, product performance, customer spending behavior, and the impact of discounts once cleaned.

#### **HANDLING MISSING VALUES**

```python
print(df.isnull().sum())
```

######

```python
cat_cols = df.select_dtypes(include=['object']).columns

for col in cat_cols:
     df[col].fillna(df[col].mode()[0])
```

```python
float_cols = df.select_dtypes(include=['float64']).columns

for col in float_cols:
    df[col] = df[col].fillna(df[col].median())
```

#### 9. MONTH-WISE SALES TREND

#### To analyze and visualize how total sales change over time by calculating monthly revenue and plotting it as a line chart.

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])

monthly_sales = df.groupby(df['Transaction Date'].dt.to_period('M'))['Total Spent'].sum()

plt.figure(figsize=(12,5))
monthly_sales.plot(kind='line', marker='o', color='darkred')
plt.title("Monthly Sales Trend", fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.show()
```

#### *I N S I G H T :-*

1. Sales show a repeating up-and-down pattern every year, indicating seasonal trends.

2. Revenue usually increases around mid-year and end-year, suggesting peak shopping seasons.

3. Early 2022 had the highest monthly sales, while the end of 2024 shows another strong rise.

4. Some months show dips, meaning low-demand periods or fewer promotions.

5. Overall, sales remain stable across years, with no major long-term decline.
